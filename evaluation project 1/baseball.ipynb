```python
# importing libraries to calculate the variance inflation factor, which may result in low accuracy
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(x1):
    vif=pd.DataFrame()
    vif["Variables"]=x1.columns
    vif["VIF factor"]=[variance_inflation_factor(x1.values,i) for i in range(x1.shape[1])]
    return (vif)
```

```python
x=dfn.iloc[:,1:]
y=dfn.iloc[:,0]
```

```python
calc_vif(x)
```

vif is criticallly high for the columns R,AB,H. AND RA,ER,ERA

observation : WE WILL DROP ER and ERA BEACUSE VIF IS MAXIMUM FOR IT AND 3 COLUMNS RA,ER AND ERA ARE MORE OR LESS HAVE STRONG NEGATIVE CORRELATION

observation: still it can be seen that multicollinearity exists between AB and H, but we will drop AB because H is positively correlated with the target variable.

# splitting of train and test data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(dfx,y,test_size=0.20,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

for determining the accurate random state

```python
def model_selection(instance):
    maxacc=0
    rs=0
    for i in range(0,100):
        x_train,x_test,y_train,y_test=train_test_split(dfx,y,random_state=i,test_size=0.30)
        instance.fit(x_train,y_train)
        pred_train=instance.predict(x_train)
        pred_test=instance.predict(x_test)
        if((r2_score(y_test,pred_test))>maxacc):
            maxacc=r2_score(y_test,pred_test)
            rs=i
        print(f"at random state {i}, training accuracy is {r2_score(y_train,pred_train)}")
        print(f"at random state {i}, testing accuracy is {r2_score(y_test,pred_test)}")
        print(f"at random state {i}, mean squared error is {mean_squared_error(y_test,pred_test)}")
        print(f"at random state {i}, mean absolute error is {mean_absolute_error(y_test,pred_test)}")
        print("\n")
    print("Max accuracy at random state",rs, "=",maxacc)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
model_selection(knr)
```

# building the model with KNeighbors regressor

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor(algorithm='auto', leaf_size= 30, n_neighbors= 6, weights='uniform')
x_train,x_test,y_train,y_test=train_test_split(dfx,y,random_state=29,test_size=0.30)
instance=knr
instance.fit(x_train,y_train)
pred_train=instance.predict(x_train)
pred_test=instance.predict(x_test)

print(f"at random state {29} testing accuracy is {r2_score(y_test,pred_test)}")
print(f"at random state {29} mean squared error is {mean_squared_error(y_test,pred_test)}")
print(f"at random state {29}, mean absolute error is {mean_absolute_error(y_test,pred_test)}")
```
